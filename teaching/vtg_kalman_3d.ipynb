```python
import csv
import matplotlib.pyplot as plt

# Arrays of data

# IMU
imutime = []
ax = []
ay = []
az = []
gx = []
gy = []
gz = []
mx = []
my = []
mz = []

# GPS
gpstime = []
lat = []
lon = []
alt = []
```

```python
# parse csv data

with open('vtg_log0.csv', 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        if row[1] == '2':
            imutime.append(float(row[0]) / 1000.0)
            ax.append(float(row[2]))
            ay.append(float(row[3]))
            az.append(float(row[4]))
            gx.append(float(row[5]))
            gy.append(float(row[6]))
            gz.append(float(row[7]))
            mx.append(float(row[8]))
            my.append(float(row[9]))
            mz.append(float(row[10]))
        elif row[1] == '1':
            gpstime.append(float(row[0]) / 1000.0)
            lon.append(float(row[2]))
            lat.append(float(row[3]))
            alt.append(float(row[4]))
```

```python
# plot things
plt.plot(imutime, ax, 'r', imutime, ay, 'b', imutime, az, 'g')

plt.xlabel("Time (s)")
plt.ylabel("Acceleration (m/s^2)")
plt.legend(['x', 'y', 'z'])
plt.show()

```
